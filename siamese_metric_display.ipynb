```python
import pickle

# Load metric dictionary from saved file
with open('statistical_processing/0.6_thres_115samples.pkl', 'rb') as f:
    loaded_metric = pickle.load(f)

print(loaded_metric)
```

```python
total_tp, total_tn, total_fp, total_fn = 0, 0, 0, 0

for name, results in loaded_metric.items():
    tp = results.count('True Positive')
    tn = results.count('True Negative')
    fp = results.count('False Positive')
    fn = results.count('False Negative')
    
    total_tp += tp
    total_tn += tn
    total_fp += fp
    total_fn += fn
    
    print(f"{name}:\nTrue Positive\t= {tp}\nTrue Negative\t= {tn}\nFalse Positive\t= {fp}\nFalse Negative\t= {fn}\n")

pred_count = total_tp + total_tn + total_fp + total_fn
print(f"\n{pred_count} Predictions")
print(f"Total:\tTP = {total_tp}, TN = {total_tn}, FP = {total_fp}, FN = {total_fn}")
```
